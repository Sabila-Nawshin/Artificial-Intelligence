```python
def KNN(X,y,test_X,K):

    
     ###calculating distances from test x point to all other points
     
    distance = np.zeros([X.shape[0],2],dtype=float)

        #no. of rows mean no. of co-ordinates
        #the two columns mean the distance and the second column is the corresponding label
    i=0    

    #print(distance)

    while i < X.shape[0]:
        x2 = m.pow( X[i,0] - test_X[0,0] , 2)
        #x2 = int(x2)
        y2 = m.pow( X[i,1] - test_X[0,1] , 2)
        #y2 = int(y2)
        d = m.sqrt(x2+y2)
        #d = int(d)
            
        distance[i,0] = d
        distance[i,1] = float(y[i])
            
        i = i + 1

            #######################################

            #sort
    distance = distance[np.argsort(distance[:,0])]  
            
    label = 0.0
        
    for i in range(K):
        label+= distance[i,1]
                

   
    return label



```
